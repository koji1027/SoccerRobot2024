```python
import math
import matplotlib.pyplot as plt

Rs = 5.2 #Ω ゲート抵抗
ciss = 926 #pF
Qg = 2.7 #nC
Vg = 12 #V
fsw = 20 #kHz
res = 1000 #PWMの分解能
Tth = 1 / fsw / 10 ** 3 / res * 10 ** 6 #us

def cal_time(Rg):
    Ig_max = Vg / (Rg + Rs)
    ig_3tau = Ig_max * (1 - 0.95)
    t = (-ciss * 10 ** -6 * (Rg + Rs)) * math.log(ig_3tau * (Rg + Rs) / Vg)
    return t #us

R = []
t = []
for i in range(1, 101):
    R.append(i)
    t.append(cal_time(i))

plt.plot(R, t, color="Blue")
plt.plot(R, [Tth] * len(R), color="Red")
plt.show()

print(cal_time(10), Tth)
```
